# Evaluate LIANA on the THA scenario

This notebook walks through the simplest SpaCCBench workflow: loading a
scenario, evaluating a bundled reference adapter, and reading the D1-D4
numbers.


```python
from spaccbench import evaluate, list_methods, list_scenarios
```

## What's available?

```python
for m in list_methods():
    marker = '✓' if m['status'] == 'bundled' else '·'
    print(f"  {marker}  {m['name']}  ({m['status']})")
```

```python
for s in list_scenarios():
    print(s)
```

## Run the four-dimensional evaluation

```python
result = evaluate(method='LIANA', scenario='tha')
```

```python
print(f"method   = {result['method']}")
print(f"scenario = {result['scenario']}")
print(f"D1 hit rate     : {result['d1']['fraction']:.2%}")
print(f"D2 Pearson r    : {result['d2']['pearson_r']:.3f}")
print(f"D2 cosine       : {result['d2']['cosine']:.3f}")
print(f"D2 JS divergence: {result['d2']['js']:.3f}")
print(f"D3 Moran's I    : {result['d3']['morans_i']:.3f}")
print(f"D3 Geary's C    : {result['d3']['gearys_c']:.3f}")
print(f"D4 mean AUC     : {result['d4']['mean_auc']:.3f}")
print(f"D4 perm p-value : {result['d4']['perm_p']:.3g}")
```

## Cohort-level ranking

Combine LIANA and COMMOT to see a rank-score composite.

```python
from spaccbench import compose_cohort

results = [
    evaluate('LIANA', 'tha'),
    evaluate('COMMOT', 'tha'),
]
table = compose_cohort(results)
table[['d1_fraction', 'd2_pearson_r', 'd3_morans_i', 'd4_mean_auc', 'composite_geo']]
```
